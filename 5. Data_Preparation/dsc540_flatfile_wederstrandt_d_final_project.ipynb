## 2. Format data into a more readable format

##### I have created a method safe_convert_to_int64 to convert three number columns into int64. 
##### Changed the type for Analysis_Type to  category and Assessment_Point to a float or NaN for errors


```python
# PTSDpubs_ID, Diag_Change_Def_Num, Clin_Meaningful_Def_Num convert to Int64
# Function to convert to Int64, handling potential errors
def safe_convert_to_int64(series):
    try:
        # Attempt to convert directly
        return series.astype('Int64')
    except ValueError:
        # If direct conversion fails, try to convert non-numeric to NaN first
        return pd.to_numeric(series, errors='coerce').astype('Int64')

# Apply conversion
outcomes_cleaned['PTSDpubs_ID'] = safe_convert_to_int64(outcomes_cleaned['PTSDpubs_ID'])
outcomes_cleaned['Diag_Change_Def_Num'] = safe_convert_to_int64(outcomes_cleaned['Diag_Change_Def_Num'])
outcomes_cleaned['Clin_Meaningful_Def_Num'] = safe_convert_to_int64(outcomes_cleaned['Clin_Meaningful_Def_Num'])
```

```python
# Analysis_Type, Interv_Group, Missing_Data_Method, Stats_Analysis_Method convert to category
outcomes_cleaned['Analysis_Type'] = outcomes_cleaned['Analysis_Type'].astype('category')
```

```python
# Assessment_Point convert to float64 by coercing non-num to NaN
outcomes_cleaned['Assessment_Point'] = pd.to_numeric(outcomes_cleaned['Assessment_Point'], errors='coerce')
```

# 3. Identify outliers using the IQR method

##### I created a method that returns a dataframe of whether or not an outlier exists.
##### I used the results of this dataframe to remove the outliers from the outcommes dataframe

```python
# Identify outliers using the IQR method
def detect_outliers_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 -q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (data < lower_bound) | (data > upper_bound)
```

```python
## Here, I created a new list of the numeric columns since the 

# Create a list of numeric columns
num_cols = [
    'Percent_Diag_Change',
    'Diag_Change_Numerator',
    'Diag_Change_Denominator',
    'Pcnt_Clin_Meaningful_Resp',
    'Clin_Meaningful_Numerator',
    'Clin_Meaningful_Denominator',
]

# Convert all non-numeric to NaN
for col in num_cols:
    outcomes_cleaned[col] = pd.to_numeric(outcomes_cleaned[col], errors='coerce')
```

```python
outliers_df = pd.DataFrame()

for col in num_cols:
    outliers_df[col + '_outlier'] = detect_outliers_iqr(outcomes_cleaned[col])
```

```python
# Identify bad data based on logic checks, e.g.:
# - Percent columns should be between 0 and 1
# - Numerator should be <= Denominator
bad_data_conditions = (
    (outcomes_cleaned['Percent_Diag_Change'] < 0) | (outcomes_cleaned['Percent_Diag_Change'] > 1) |
    (outcomes_cleaned['Pcnt_Clin_Meaningful_Resp'] < 0) | (outcomes_cleaned['Pcnt_Clin_Meaningful_Resp'] > 1) |
    (outcomes_cleaned['Clin_Meaningful_Numerator'] > outcomes_cleaned['Clin_Meaningful_Denominator']) |
    (outcomes_cleaned['Diag_Change_Numerator'] > outcomes_cleaned['Diag_Change_Denominator'])
)

outcomes_cleaned['bad_data'] = bad_data_conditions
```

```python
# Combine outlier flags and bad data flags
outcomes_cleaned['any_outliers'] = outliers_df.any(axis=1)
outcomes_cleaned['potential_issues'] = outcomes_cleaned.any_outliers | outcomes_cleaned.bad_data

problem_rows = outcomes_cleaned[outcomes_cleaned.potential_issues]
```

## 5. Conduct Fuzzy Matching

##### Fuzzy matching compares each treatment name in the dataset to a list of standard treatment names using a similarity score, rather than requiring exact matches. 
##### It selects the closest match when the similarity exceeds a defined threshold, allowing correction of inconsistently spelled or formatted values like "cbt", 
##### "C.B.T.", or "cognitive bt" to a standardized label such as "Cognitive Behavioral Therapy."

```python
# Unique treatment values in which to match against
t = outcomes_cleaned_no_outliers_df['Treatment_Name'].dropna().unique()
```

```python
# Define a map of similar values

def fuzzy_match_choices(value, choices, threshold=90):
    match, score, _ = process.extractOne(value, choices, scorer=fuzz.ratio)
    return match if score >= threshold else value
```

```python
# Create a cleaned version of the column Treatment_Name

outcomes_cleaned_no_outliers_df['Treatment_Name_Cleaned'] = outcomes_cleaned_no_outliers_df['Treatment_Name'].apply(lambda x: fuzzy_match_choices(x, t))
outcomes_cleaned_no_outliers_df
```

```python
# Create a list of standard labels
standard_categories = [
    "CBT", "CPT", "PE", "EMDR", "Supportive Counseling", "Mindfulness",
    "Medication", "Exposure Therapy", "Yoga", "DBT", "NET",
    "Waitlist", "Psychoeducation", "TAU", "IPT", "Relaxation Therapy",
    "Behavior Therapy", "Cognitive Therapy"
]
```

```python
# This function maps a treatment name to the closest standardized treatment category
# using fuzzy string matching. It compares each input value to a predefined list of
# standard treatment categories and returns the closest match if the similarity score
# exceeds the threshold. If no match meets the threshold, it assigns the category "Other".

def fuzzy_map_treatment(value, standard_list, threshold=80):
    match, score, _ = process.extractOne(value, standard_list, scorer=fuzz.token_sort_ratio)
    return match if score >= threshold else 'Other'
```

```python
# Fill Treatment_Category from Treatment_Name mapping
outcomes_cleaned_no_outliers_df['Treatment_Category'] = outcomes_cleaned_no_outliers_df['Treatment_Name'].apply(
    lambda x: fuzzy_map_treatment(x, standard_categories))
```
